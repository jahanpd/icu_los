```r
# path_value <- Sys.getenv("ICUDATA")
path_value <- "~/Sync/icu-los/2250_APD.csv"
print(path_value)

# Load a CSV file into a data frame
data <- read.csv(path_value)
data[data == 'NULL'] <- NA
head(data)
```

```r
# create empty dataframe to mimic Brazilian DD
n <- nrow(data)
base <- data.frame(
  patientID = data$PatientID, 
  Age = as.numeric(data$AGE), # from AGE
  Gender = as.factor(data$SEX), # from SEX
  BMI = vector(mode = "numeric", length = n), # custom
  isReadmission = vector(mode = "logical", length = n), # custom
  UnitLengthStay = as.numeric(data$ICU_HRS), # from ICU_HRS
  AdmissionSourceName = as.factor(data$ICU_SRCE), # from ICU_SRCE
  Frailty = vector(mode = "logical", length = n), # custom 
  DeathProbability = as.numeric(data$ANZRODRiskOfDeath), # ANZROD
  SofaScore = as.numeric(data$SOFA), # SOFA
  AdmissionMainDiagnosisName = vector(mode = "character", length = n), # custom 
  IsChfNyhaClass4 = as.factor(data$CHR_CVS), # from chr_cvs
  IsCrfDialysis = as.factor(data$CHR_REN), # from chr_ren
  IsCirrhosisChildC = vector(mode = "logical", length = n), # custom
  IsHepaticFailure = as.factor(data$HEPFAIL), # from hep_fail
  IsSolidTumorMetastatic = as.factor(data$METAST), # from METAST
  IsHematologicalMalignancy = vector(mode = "logical", length = n), # custom
  IsImmunosupression = vector(mode = "logical", length = n), # custom
  IsSevereCopd = as.factor(data$CHR_RESP), # chr_resp
  IsImmunosuppressionRx = vector(mode = "logical", length = n), # custom
  IsAids = as.factor(data$AIDS), # from AIDS
  IsDiabetes = as.factor(data$DIABETES), # from diabetes
  IsMechanicalVentilation = vector(mode = "logical", length = n), # custom
  IsNonInvasiveVentilation = as.factor(data$NIV_IND), # from NIV_IND
  IsVasopressors = as.factor(data$INOTROP_IND), # from inotrop_ind
  IsAcuteKidneyInjury = vector(mode = "logical", length = n), # custom
  IsRenalReplacementTherapy = as.factor(data$RENAL_IND), # from RENAL_IND
  IsNeutropenia = vector(mode = "logical", length = n), # custom
  LowestSystolicBloodPressure24h = as.numeric(data$SYSTOLICLO), # systoliclo
  LowestDiastolicBloodPressure24h = as.numeric(data$DIASTOLICLO), # diastoliclo
  LowestMeanArterialPressure24h = as.numeric(data$MAPLO), # maplo
  LowestGlasgowComaScale24h = as.numeric(data$GCS), # gcs
  LowestPlateletsCount24h = as.numeric(data$PLATLO), # platlo
  pH24h = as.numeric(data$PH_ANZ), # ph_anz
  PaO24h = as.numeric(data$PAO2_ANZ), # pao2_anz
  PaCO21h = as.numeric(data$PACO2_ANZ), # paco2_anz
  FiO224h = as.numeric(data$FIO2_ANZ), # fio2_anz
  HighestHeartRate24h = as.numeric(data$HRHI), # hrhi
  HighestRespiratoryRate24h = as.numeric(data$RRHI), # rrhi
  HighestTemperature24h = as.numeric(data$TEMPHI), # temphi
  HighestLeukocyteCount24h = as.numeric(data$WCCHI), # wcchi
  HighestCreatinine24h = as.numeric(data$CREATHI), # creathi
  Bilirubin24h = as.numeric(data$BILI_ANZ), # bili_anz
  Lactate24h = as.numeric(data$LACTATE), # lactate
  Urea24h = as.numeric(data$UREA_ANZ), # urea_anz
  stringsAsFactors = FALSE
)
head(base)
```

# New/Custom Variables

| Name     | Description |
| ----------- | ----------- |
| BMI      | standardly derived from height and weight       |
| isReadmission   | combined admepisode and readmitted        |
| Frailty   |  from clinical frailty score cutoff       |
| DeathProbability   |  calculated from ANZROD rather than SAPS3      |
| AdmissionMainDiagnosisName   | string extracted  from ap3 diag and subcode lookup     |
| AdmissionMainDiagnosisName   | string extracted  from ap3 diag and subcode lookup     |
| IsCirrhosisChildC   |  from cirrhosis and chr_liv combined    |
| IsHematologicalMalignancy |  from leukaem and lymphoma combined    |
| IsImmunossupression |  from immundis and immunsup combined    |
| IsImmunossupressionRx |  from  immunorx   |
| IsMechanicalVentilation | combination of "INV_IND", "INV_DAYONE", "INTUBATED", "VENTILATED", "INV_HOURS" |
| IsAcuteKidneyInjury |  from  UO and Cr definitions or ?ARF although apparently bad definition  |
| IsNeutropenia |  from   wcchi <1 OR wcclo <1  |

# Data preparation
The following cells prepare columns as per the new/custom variables

```r
# calculate BMI
base$BMI <- as.numeric(data$WEIGHT) / (as.numeric(data$HEIGHT) / 100)**2
mean(base$BMI, na.rm = TRUE)
```

```r
# "admepisode>1 is a readmission episode to ICU
# readmitted=1 is a readmission episode to ICU"
base$isReadmission <- data$AdmEpisode > 1 | data$READMITTED == 1
print(sum(base$isReadmission) / n)
```

```r
base$Frailty <- as.numeric(data$FRAILTY) > 4
print(sum(base$Frailty, na.rm = TRUE) / n)
```

```r
base$IsCirrhosisChildC <- data$CIRRHOS == 1 | data$CHR_LIV == 1
print(sum(base$IsCirrhosisChildC, na.rm = TRUE)/n)
```

```r
base$IsHematologicalMalignancy <- data$LEUKAEM == 1 | data$LYMPHOMA == 1
print(sum(base$IsHematologicalMalignancy, na.rm = TRUE)/n)
```

```r
base$IsImmunosupression <- data$IMMUNDIS == 1 | data$IMMUNSUP == 1
print(sum(base$IsImmunosupression, na.rm = TRUE)/n)
```

```r
base$IsImmunosuppressionRx <- data$IMMUNRX == 1
print(sum(base$IsImmunosuppressionRx, na.rm = TRUE)/n)
```

```r
# "INV_IND", "INV_DAYONE", "INTUBATED", "VENTILATED", "INV_HOURS"
base$IsMechanicalVentilation24h <- data$INV_DAYONE == 1
base$IsMechanicalVentilation <- data$INV_IND == 1 | data$INV_DAYONE == 1 | data$INTUBATED == 1 | data$VENTILATED == 1 | as.numeric(data$INV_HOURS) > 0
print(sum(base$IsMechanicalVentilation24h, na.rm = TRUE)/n)
print(sum(base$IsMechanicalVentilation, na.rm = TRUE)/n)
```

```r
base$IsAcuteKidneyInjury <- as.numeric(data$ARF) == 1
print(sum(base$IsAcuteKidneyInjury, na.rm = TRUE)/n)
```

```r
base$IsNeutropenia <- data$WCCHI < 1 | data$WCCLO < 1
print(sum(base$IsNeutropenia, na.rm = TRUE)/n)
```
